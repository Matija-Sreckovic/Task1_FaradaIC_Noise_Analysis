# 1. Noise Analysis

In this notebook, I will analyze the dataset corresponding to the first task - the noise analysis for sensors.

## 1.0. Loading the data

In the next cell, I read the data - there are 120 (15x8) DataFrames in total.

Note that the first two rows contain miscellaneous information - the date and time of the measurement, MFC1, MFC2, time, h...

**Each DataFrame has two columns - voltage ('V') and current (measured in 'uA', which is also the name of the second column).**

```python
import pandas as pd

sensor_dataframes = []
date_times = []
MFC_t_h_list = []

for i in range(8):
    sensor_dataframes.append([])
    date_times.append([])
    MFC_t_h_list.append([])
    
for i in range(8):
    for j in range(15):
        # Define the file path
        file_path = f"20230614_CA_noise_test_chamberopen_with pins_0_0_0_{j}_{i}.csv"
        
        # Open the file and read the first two lines
        with open(file_path, "r", encoding="utf-8") as file:
            date_times[i].append(file.readline().strip())
            MFC_t_h_list[i].append(file.readline().strip())
            
        measurement = pd.read_csv(file_path, sep=';', skiprows=2)
        measurement.columns = ['V', 'uA']
        sensor_dataframes[i].append(measurement)

sensor_dataframes
```

## 1.1. A First Look: Plotting & Informal Analysis

To get familiar with the dataset, I plot each DataFrame and take a quick, informal look at whether there are any channels that behave differently than others.

```python
#Now let's plot graphs for each sensor!
import matplotlib.pyplot as plt

for i in range(8):
    print(f"SENSOR {i}:")
    for j in range(15):
        plt.figure()  # Create a new figure for each plot
        plt.plot(sensor_dataframes[i][j]['V'], sensor_dataframes[i][j]['uA'])
        plt.title(f"Sensor {i}, Measurement {j}")
        plt.xlabel("V")
        plt.ylabel("uA")
        plt.show()  # Show the plot separately
```

We see from the plots that:
- sensor 3 exhibits slightly different behavior in the first 4 measurements than the others (negative current as voltage goes to 0)
- **sensor 5 behaves rather erratically, with much higher noise than the others.**

The second observation is much more significant!

```python
# To look at 8 plots instead of 120, we can compute the mean current across measurements for each sensor.

voltages = sensor_dataframes[i][j]['V'].to_list()
mean_dataframes = []
for i in range(8):

    # Start by extracting the voltage column from the first DataFrame
    result_df = pd.DataFrame()
    result_df["V"] = sensor_dataframes[i][0]["V"]
    
    # Add each measurement's current ('uA') as a separate column
    for j, df in enumerate(sensor_dataframes[i]):
        result_df[f"uA_{j}"] = df["uA"]
    
    result_df["Mean_uA"] = result_df.iloc[:, 1:].mean(axis=1)  # Mean across all 'uA' columns
    mean_df = result_df[['V', 'Mean_uA']]
    mean_dataframes.append(mean_df)
            
    # Display the result
    plt.figure()
    plt.plot(mean_dataframes[i]['V'], mean_dataframes[i]['Mean_uA'])
    plt.title(f"Sensor {i} - Mean Signal")
    plt.xlabel("V")
    plt.ylabel("uA")
    plt.show()  # Show the plot separately
```

```python
# Let's now ignore the first 50 rows to see if there are any subtleties...
for i in range(8):
    mean_dataframe_after_51_row = mean_dataframes[i].iloc[51:]
    # Display the result
    plt.figure()
    plt.plot(mean_dataframe_after_51_row['V'], mean_dataframe_after_51_row['Mean_uA'])
    plt.title(f"Sensor {i} - Mean, Ignoring the first 50 Rows")
    plt.xlabel("V")
    plt.ylabel("uA")
    plt.show()  # Show the plot separately
```

### Observation:

- Sensors 0, 1 and 3 require a slightly higher voltage to exhibit "random noise" behavior.
- Sensors 2, 4, 6 and 7 normalize faster.
- Sensor 5's amplitude is about 100 greater than the other sensors'.

## 1.2. Quantitative Noise Analysis

I begin a more formal analysis by computing two numerical features of each sensor:
- the **standard deviation** of the current;
- the **signal-to-noise ratio (SNR)**.

```python
noise_std_list = []
for i in range(8):
    noise_std_list.append([])
    for j in range(15):
        noise_std_list[i].append(sensor_dataframes[i][j]["uA"].std())

for i in range(8):
    print(f"Standard deviations for each measurement of Channel {i}:")
    print(noise_std_list[i])
    print("------")
    print(f"Mean standard deviation for measurements of Channel {i}:")
    print(sum(noise_std_list[i])/len(noise_std_list[i]))
    print("------")
    print(f"Standard deviation of the *averaged out* signal for Channel {i}:")
    print(mean_dataframes[i]['Mean_uA'].std())
    print("-------------------------------------------\n")
```

NB: the **measurements** of each channel are **quite stable** - even the corrupted one (no. 5). Therefore, to work with 8 numbers instead of 120, **I average out the signals and then compute one signal-to-noise ratio for each channel.**

```python
#Next, I compute the Signal-to-Noise ratio:
#First I get the mean signal across each measurement
for i in range(8):
    print(f"Channel {i} SNR: {mean_dataframes[i]['Mean_uA'].mean()/mean_dataframes[i]['Mean_uA'].std()}")
```

The noise generally outweighs the signal even in the corrupted channel 5. Note that channel 3 has a slightly smaller amplitude than the rest, and a similar mean current, so its SNR is a bit higher.

## 1.3. Outlier Detection & Corruption Confirmation

Here I compute the *$Z$-score of the standard deviations* to prove that channel 5 is indeed an outlier.

The **$Z$-score** of a datapoint is computed as follows: $$Z = \frac{X - \mu}{\sigma},$$where $X$ is the value at the datapoint, $\mu$ is the mean value of the points, and $\sigma$ is their standard deviation.

If the aboslute value of the $Z$-score of a point is greater than $3$, that point is considered a serious outlier.

```python
averaged_out_std_devs = []
z_scores = []
for i in range(8):
    averaged_out_std_devs.append(mean_dataframes[i]['Mean_uA'].std())

mu = sum(averaged_out_std_devs)/len(averaged_out_std_devs)
sigma = pd.Series(averaged_out_std_devs).std()

for i in range(8):
    z_scores.append((averaged_out_std_devs[i] - mu)/sigma)
    print(f"Z-score for Channel {i}: {z_scores[i]}")
```
